**Аналіз:** Понад половина опитаних фахівців працюють з Python, що робить цю мову ключовим інструментом у вибірці. Це пояснює високий попит на Python-розробників та аналітиків даних.

```python
python_devs_with_salary = df[(df['worked_with_python'] == True) & (df['ConvertedCompYearly'].notna())]

py_progers_by_country = python_devs_with_salary.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).round(1)

py_progers_by_country.columns = ['Середня компенсація', 'Медіанна компенсація']

py_progers_by_country = py_progers_by_country.sort_values(by='Медіанна компенсація', ascending=False)

py_progers_by_country.head(20)
```
